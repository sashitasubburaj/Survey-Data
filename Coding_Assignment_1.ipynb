I wanted to see where most of the gpas of the students I surveyed was on the 4.0 scale. I was pleasantly surprised to see that so many students had such high gpas!

```python
from scipy.stats import pearsonr

movies = survey_df["Amount of Movies Watched in Past Month: "]

student_gpa = survey_df["GPA: "]

p = pearsonr(student_gpa,movies)
print (p[0])
```

Through this project, something I wanted to explore was the potential impact media consumption via movies may have on academic performance. By calculating the r coefficient, I could make inferences on whether these two variables were correlated. Because the coefficient is so far away from 1, there doesn't seem to be much of a relationship that exists here.

```python

plt.scatter(student_gpa,movies, marker="*")

plt.xlabel("Student GPA on 4.0 Scale")
plt.ylabel("Movies Watched in Last Month")

plt.title("Movies Watched vs Academic Performance")
```
